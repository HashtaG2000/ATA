```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("/content/Telco.csv")

# Select relevant columns for clustering
X = data[['Satisfaction Score', 'Tenure in Months']]

# Specify the number of clusters
k = 3

# Initialize GaussianMixture
gmm = GaussianMixture(n_components=k)

# Fit GaussianMixture to the data
gmm.fit(X)

# Get cluster labels
labels_gmm = gmm.predict(X)

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels_gmm)

# Add cluster labels to the original dataframe
data['Cluster (GMM)'] = labels_gmm

# Visualize the clustering
plt.figure(figsize=(10, 6))
plt.scatter(data['Satisfaction Score'], data['Tenure in Months'], c=labels_gmm, cmap='viridis', alpha=0.5)
plt.xlabel('Satisfaction Score')
plt.ylabel('Tenure in Months')
plt.title('Gaussian Mixture Model Clustering of Satisfaction Score vs Tenure in Months')
plt.show()

# Print silhouette score
print("Silhouette Score:", silhouette_avg)
```
