```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("/content/Telco.csv")

# Select relevant columns for clustering
X = data[['Satisfaction Score', 'Tenure in Months']]

# Specify the number of clusters
k = 3

# Initialize KMeans
kmeans = KMeans(n_clusters=k)

# Fit KMeans to the data
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)

# Add cluster labels to the original dataframe
data['Cluster'] = labels

# Print silhouette score
print("Silhouette Score:", silhouette_avg)

# Visualize the clustering
plt.figure(figsize=(10, 6))
plt.scatter(data['Satisfaction Score'], data['Tenure in Months'], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', label='Centroids')
plt.xlabel('Satisfaction Score')
plt.ylabel('Tenure in Months')
plt.title('K-means Clustering of Satisfaction Score vs Tenure in Months')
plt.legend()
plt.show()
```
