```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("/content/Telco.csv")

# Select relevant columns for clustering
X = data[['Satisfaction Score', 'Tenure in Months']]

# Specify the number of clusters
k = 3

# Initialize AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=k)

# Fit AgglomerativeClustering to the data
agg_clustering.fit(X)

# Get cluster labels
labels_agg = agg_clustering.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels_agg)

# Add cluster labels to the original dataframe
data['Cluster (Agglomerative)'] = labels_agg

# Visualize the clustering
plt.figure(figsize=(10, 6))
plt.scatter(data['Satisfaction Score'], data['Tenure in Months'], c=labels_agg, cmap='viridis', alpha=0.5)
plt.xlabel('Satisfaction Score')
plt.ylabel('Tenure in Months')
plt.title('Agglomerative Hierarchical Clustering of Satisfaction Score vs Tenure in Months')
plt.show()

# Print silhouette score
print("Silhouette Score:", silhouette_avg)
```
